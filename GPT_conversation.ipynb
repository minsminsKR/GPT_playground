```python
# The purpose is making two Chat GPT having a conversation

# a : The personality you want to give to model1
# b : The personality you want to give to model2
# c : The topic you want to talk about

!pip install openai
import openai
import time
openai.api_key = "Your API key here."

def start_chat():
    a = input("Set first model's feature : \nex)Your personality is __. You are __. You are ___...\n")
    a_name = input("Set first model's name : ")
    b = input("Set second model's feature : ")
    b_name = input("Set second model's name : ")
    c = input('Set subject or discussion : \nex)Do pets have a positive or negative effect on our lives?\n')
    n = input('How many times would you like to proceed with the conversation? : ')
    lan = input('Set language : ')

    model = "gpt-3.5-turbo"

    # Initial message from model 1
    model1 = openai.ChatCompletion.create(
        model=model,
        messages=[
            {"role": "system", "content": f"{a}. You must reply in two sentences with {lan}."},
            {"role": "user", "content": f"Let's have a discussion about {c}. Make a any question about that subject."}
        ]
    )


    # Print response from model 1
    print(f"\n{a_name} :", model1.choices[0].message.content)

    # Exchange messages n times
    for i in range(int(n)):
        time.sleep(1)
    
        # Send message from model 1 to model 2
        model2 = openai.ChatCompletion.create(
            model=model,
            messages=[
                {"role": "system", "content": f"{b}. You must reply in two sentences with {lan}."},
                {"role": "user", "content": model1.choices[0].message.content}
            ]
        )

        # Print response from model 2
        print(f"{b_name} :", model2.choices[0].message.content)
        
        time.sleep(1)
        
        # Send message from model 2 to model 1
        model1 = openai.ChatCompletion.create(
            model=model,
            messages=[
                {"role": "system", "content": f"{a}. You must reply in two sentences with {lan}."},
                {"role": "user", "content" : model2.choices[0].message.content}
            ]
        )
    
        # Print response from model 1
        print(f"{a_name} :", model1.choices[0].message.content+'\n')

start_chat()
```
