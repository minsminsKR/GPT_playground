```python
# The purpose is drawing a painting about you using Chat GPT 3.5
!pip install openai
import openai
openai.api_key = "Your API key here."

def painting_gpt():
    a = input("Let me know your MBTI or personality : ")
    a_name = input("Let me know your name : ")

    model = "gpt-3.5-turbo"
    model = openai.ChatCompletion.create(
        model = model,
        messages=[
            {'role': 'system', 'content': f'You are a great artist. You should reply in one sentence, only keywords'},
            {'role' : 'user', 'content': f'I want to draw a painting. The subject of painting is {a_name} and {a}. Show me 10 keywords about it except name.'}
        ]
    )

    print(f"\n{a_name} :", model.choices[0].message.content, "\nDrawing...")
    print(
    openai.Image.create(
        prompt = model.choices[0].message.content,
        n=1,
        size="256x256"
    )
    )

painting_gpt()
```
